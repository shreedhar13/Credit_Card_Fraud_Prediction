# Under Samppling

Build a sample dataset containing similar distribution of normal transaction and Fraudlent transaction

```python
#ie;bcz we have only 492 Fraudlent tranasaction data,,,so we are going to consider sample of 492 legit data points...thus to make dataset balanced..
```

```python
legit_sample= Legit.sample(n=492)#randomly selecting 492 
```

```python
#now concatinate legit_sample with fraudlent data.....thus we get balanced dataset
```

```python
new_dataset = pd.concat([legit_sample,Fraud],axis=0)#0 means row by row operation is done,,,ie;one by one row is added..
```

# NATURE of dataset is almost same....

```python
X = new_dataset.iloc[:,:-1] #or new_dataset.drop(columns='Class',axis=1)  droping column so 1.. 
y = new_dataset.iloc[:,-1]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.1,stratify=y,random_state=2)
#startify=y -> evenely split the legit and fraud or we can say target feature values are evenly splitted
#if i dont use it then,,,,,X_train may have more no of Legit data compare to X_test or vice versa....
```

```python
X_train
```

# model training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)#accuracy on training data
```

```python
model.score(X_test,y_test)#testing accuracy
```

```python
#OR   testing accuracy
y_hat = model.predict(X_test)
accuracy_score(y_hat,y_test)
```
